```python
# Multi-input Class Average Program

def main():
    print("Enter all grades separated by commas (e.g., 60, 70, 80):")
    user_input = input("Grades: ")

    # 1. Split the string by commas
    # 2. Convert each piece to a float
    try:
        grades = [float(g.strip()) for g in user_input.split(",")]

        if len(grades) > 0:
            average = sum(grades) / len(grades)
            print("-" * 30)
            print(f"Total grades entered: {len(grades)}")
            print(f"The class average is: {average:.2f}")
    except ValueError:
        print("Error: Please make sure you only enter numbers separated by commas.")

if __name__ == "__main__":
    main()
```
